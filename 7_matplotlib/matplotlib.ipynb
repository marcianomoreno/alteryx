# Matplotlib

Matplotlib grafica los datos en figuras (ventanas, widgets de Jupyter, etc.), cada una puede contener uno o más `Axes`, un área en donde los puntos pueden especificarse en coordenadas x-y y otras dimensiones. la forma más sencilla de crear una figura con axes es por mediodo de `pypplot.subplots`. `Axes.plot` se emplea par dibujar datos en los axes.

Los elementos de las gráficas de Matplotlib incluyen:

| Elemento | Descripción |
| --- | --- |
| Figure | La figura completa. Incluye los axes (gráficas) y el canvas. |
| Axes | Región en la figura con la gráfica |
| Axis | Determinan los límites de las gráficas y generan ticks. |
| Artist | Cualquier elemento visible en la gráfica. Se dibujan en el canvas. |

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4], [1,4,2,3])
```

El método `plot` permite graficar en el axes vigente.

```python
plt.plot([1,2,3,4], [1,4,2,3])
```

Es posible crear una figura con más de una gráfica.

```python
fig = plt.figure()
fig, ax = plt.subplots()
fig, axs = plt.subplots(2,2)
```

# Las interfaces de Matplotlib
| Interfaz | Comentarios |
| --- | --- |
| Orientada a objetos | Permite crear explícitamente las figuras y axes e invocar métodos en ellos. Recomendada para funciones o scritps que serán reusados en otro proyecto. |
| pyplot | Crea gráficas automáticamente. Recomendada para graficación interactiva (p.e. en Jupyter). |

Interfaz orientada a objetos

```python
x = np.linspace(0, 2, 100)
fig, ax = plt.subplots()
ax.plot(x, x, label="linear")
ax.plot(x, x**2, label="quadratic")
ax.plot(x, x**3, label="cubic")
ax.set_xlabel("x label")
ax.set_ylabel("y label")
ax.set_title("Simple plot")
ax.legend()
```

Interfaz pyplot

```python
x = np.linspace(0, 2, 100)
plt.plot(x, x, label="linear")
plt.plot(x, x**x, label="quadratic")
plt.plot(x, x**3, label="cubic")
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("Simple plot")
```

La siguiente función puede emplearse para crear múltiples gráficas con distintos parámetros.

```python
def my_plotter(ax, data1, data2, param_dict):
    '''
    Helper function to make a graph
    
    Parameters
    ----------
    ax : Axes
        The axes to draw to
    
    data1 : array
        The x data
    
    data2 : array
        The y data

    param_dict : dict
        Dictionary of kwargs to pass to ax plot

    Returns
    -------
    out : list
        list of arists added
    '''

    out = ax.plot(data1, data2, **param_dict)
    return out
```

Ejemplo de uso de la función de graficación:

```python
data1, data2, data3, data4 = np.random.randn(4, 100)
fig, ax = plt.subplots(1, 1)
my_plotter(ax, data1, data2, {"marker": "x"})
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
my_plotter(ax1, data1, data2, {"marker": "x"})
my_plotter(ax2, data3, data4, {"marker": "o"})
```
